```python
import numpy as np
#sample data
math = np.array([78,92,85,61,72,95,88,54,67,100])
science = np.array([85,88,72,91,60,79,100,68,54,82])
english=np.array([70,95,81,58,90,63,77,86,92,74])
```

```python
print("Math avg:",np.mean(math))
print("Science avg:",np.mean(science))
print("English avg:",np.mean(english))
```

```python
mathHighest,mathLowest = np.max(math),np.min(math)
scienceHighest,scienceLowest = np.max(science),np.min(science)
englishHighest,englishLowest = np.max(english),np.min(english)
print("Math min and max:",mathHighest,mathLowest)
print("Science min and max:",scienceHighest,scienceLowest)
print("English min and max:",englishHighest,englishLowest)
```

```python
overAllStd=len(math);
mathPass = np.sum(math>=70)
mathFail = (overAllStd - mathPass)
sciencePass  = np.sum(science>=70)
scienceFail = (overAllStd - sciencePass)
englishPass  = np.sum(english>=70)
englishFail = (overAllStd - englishPass)
print("Count of Pass and Fail in math:",mathPass,mathFail)
print("Count of Pass and Fail in science:",sciencePass,scienceFail)
print("Count of Pass and Fail int english:",englishPass,englishFail)
```

```python
overAllRes = np.sum((math>=70) & (science>=70) & (english>=70))

print("Total pass:",overAllRes);
print("Total fail:",overAllStd - overAllRes);
```

```python
performance=np.empty(len(math),dtype=str);
print(performance);
```

```python
def assignGrade(i,avg,per):
  if avg >= 90:
    per[i] = 'A'
  elif avg >= 80 and avg < 90:
    per[i] = 'B'
  elif avg >= 70 and avg < 80:
    per[i] = 'C'
  else:
    per[i] = 'F'
```

```python
overAllAvg = np.round((math+science+english)/3,2);
print(overAllAvg);
```

```python
for i,val in enumerate(overAllAvg):
  assignGrade(i,val,performance)
print(performance)
```
