```python
#importing pandas and creating the dataframe
import pandas as pd;


data = {
    "Name":["Saad","Umair","Imran","Hashim"],
    "Roll#":[102,103,101,104],
    "Age":[21,20,22,23],
    "Address":["Kahuta","DHA","Behria","Jinnah Garden"]
}

df = pd.DataFrame(data);
print(df)
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df["Address"])
print(df[["Name","Roll#"]])
print(df.loc[0:2,"Age"])
print(df.iloc[0:,1:])
```

```python
print(df[df["Age"]>21])

print(df[df["Address"] == "Behria"])
```

```python
df["CGPA"]=[3.11,3.21,3.31,3.15]
df = df.drop("Roll#",axis=1)
print(df)
```

```python
print(df.sort_values(by="CGPA",ascending=False))
```

```python
df.loc[4]=["Hassan",23,"G-11",3.25]
print(df)
df = df.drop(0)
df=df.reset_index(drop=True)
print(df);
```

```python
newRow=pd.DataFrame({
    "Name":["Saad"],
    "Age":[21],
    "Address":["Kahuta"],
    "CGPA":[3.40]
})
df = df.drop(0);
df = pd.concat([df,newRow],ignore_index=True)
print(df)
```

```python
print(df.isnull())
print(df.isnull().sum())
```

```python
import numpy as np
df.loc[4]=["",np.nan,"",np.nan]
print(df)
```

```python
print(df.isnull().sum())
```

```python
df["CGPA"].fillna(df["CGPA"].mean())
print(df)
```

```python
df["Age"].fillna(method="ffill",inplace=True)
print(df)
```

```python
df.loc[5] = [np.nan,22,np.nan,3.22]
print(df)
```

```python
df["Name"].fillna("Unknown",inplace=True)
print(df)
```

```python
df["Address"] = df["Address"].fillna("Unknown")
print(df)
```
