```python
import pandas as pd;
```

```python
print(pd.__version__)
```

```python
data={
    "Name":["Imran","Aslam","Hasan","Irfan"],
    "Age":[22,23,21,24],
    "Salary":[20000,22000,23000,21000]
}
print(data);
```

```python
df=pd.DataFrame(data);
print(df)
```

```python
print(df["Name"])
print(df[["Age","Salary"]]);
```

```python
print(df.loc[0,"Name"]);
print(df.iloc[0,1]);
print(df.iloc[1]);
print(df.loc[0]);
```

```python
print(f"shape:{df.shape}");
print(f"columns:{df.columns}");
print(f"Data types of each column:{df.dtypes}");
```

```python
print("Detailed info:",df.describe())
```

```python
df["Bonus"] = df["Salary"] * 0.1;
print(df)
```

```python
print(df.drop("Bonus",axis=1));
```

```python
n_row={
    "Name":"Eve",
    "Age":32,
    "Salary":26
}

# df = df.append(n_row,ignore_index=True); #depreciated
print(pd.DataFrame([n_row]));
df = pd.concat([df,pd.DataFrame([n_row])],ignore_index=True);
print(df);
```

```python
print(df.drop([4,3]));
```

```python
print(df.isnull());
```
