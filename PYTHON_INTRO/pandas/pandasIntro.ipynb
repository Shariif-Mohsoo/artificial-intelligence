```python
import pandas as pd
print(pd.__version__)
```

```python
#Loading and storing datasets 
#Method 1: From Dictionary
data ={
    "Name":["Ali","Hassan","Imran","Danish"],
    "Age":[21,22,21,23],
    "CGPA":[3.88,3.87,3.89,3.90]
}
print(data) #The json will be printed.
df = pd.DataFrame(data)
print(df)#table will be created.
```

```python
#Loading and storing datasets -- Contd.
#Method 2: From lists
names=["Alice","Bob","Charlie"]
ages=[20,21,23]
df_list=pd.DataFrame({
    "Name":names,
    "Age":ages
})
print(df_list)
```

```python
# Loading external data
#can be loaded from csv,excel and sql.
#demo for excel
df_excel=pd.read_excel("Datasets /BSCS.xlsx");
print(df_excel)
```

```python
#reading data from csv file
df_csv = pd.read_csv("Datasets /BSCS.csv")
print(df_csv)
```

```python
# selecting data
#selecting columns
print(df['Name']) #selecting the single column
print(df[["Name",'Age']]) #selecting the multiple columns
```

```python
#selecting rows
print(df)
print(df.loc[1]) #selecting the row by label
print(df.iloc[0]) #selecting row by index
print(df[df["Age"]>=22]) #filtered rows
```

```python
#Exploring data
print(df.shape)#row,cols
print(df.columns)#names of columns
print(df.dtypes) #data types
print(df.info()) #detailed info

#first and last rows
print(df.head())#default first 5 rows will be selected.
print(df.head(3))
print(df.tail(3))#last 3 rows will be selected

print(df.describe()) #statistical summary
```
