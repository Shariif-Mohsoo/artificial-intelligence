```python
import numpy as np

math = np.array([78,92,85,61,72,95,88,54,67,100])
science = np.array([85,88,72,91,60,79,100,68,54,82])
english = np.array([70,95,81,58,90,63,77,86,92,74])
```

```python
avg_math = np.mean(math)
avg_science = np.mean(science)
avg_english = np.mean(english)

print("Average Grades")
print(f"Math: {avg_math}")
print(f"Science: {avg_science}")
print(f"English: {avg_english}\n")
```

```python
print("Highest and Lowest Grades:")
print(f"Math: Highest={np.max(math)}, Lowest={np.min(math)}")
print(f"Science: Highest={np.max(science)}, Lowest={np.min(science)}")
print(f"English: Highest={np.max(english)}, Lowest={np.min(english)}")
```

```python
def count_pass_fail(subject):
    passed = np.sum(subject>=70)
    failed = np.sum(subject<70)
    return passed,failed

math_pass,math_fail = count_pass_fail(math)
sci_pass,sci_fail = count_pass_fail(science)
eng_pass,eng_fail = count_pass_fail(english)

print("Pass/Fail per Subject:")
print(f"Math: Pass={math_pass}, Fail={math_fail}")
print(f"Science: Pass={sci_pass}, Fail={sci_fail}")
print(f"English: Pass={eng_pass}, Fail={eng_fail}")

```

```python
passed_all = np.sum((math >= 70) & (science >= 70) & (english>=70))
total_stds = len(math)
fail_any = total_stds - passed_all


print("Overall Results:")
print(f"Students Passed All Subjects: {passed_all}")
print(f"Students Failed Any Subject: {fail_any}")
```

```python
overall_avg = (math + science + english)/3
print(overall_avg)
performance = np.empty(len(overall_avg),dtype=str)
for i,avg in enumerate(overall_avg):
    if avg>=90:
        performance[i]='A'
    elif avg>=80:
        performance[i]='B'
    elif avg>=70:
        performance[i]='C'
    else:
        performance[i]='F'
print("Student Performance Classification")
print("Performance Array:",performance)
```
