```python
import numpy as np;
import matplotlib.pyplot as plot;
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
print(sigmoid(3))
print(sigmoid(13))
print(sigmoid(30))
```

```python
def logistic_regression_fit(x,y,lr=0.01,epochs=1000):
    n=len(x)
    B0,B1 = 0,0
    for _ in range(epochs):
        z=B0 + B1*x
        p=sigmoid(z)
        B0 -= lr * np.mean(p-y)
        B1 -= lr * np.mean((p-y)*x)
    return B0,B1
```

```python
def predict_prob(x,b0,b1):
    return sigmoid(b0+b1*x)
```

```python
def predict(x,b0,b1,threshold=0.5):
    return (sigmoid(b0+b1*x) >= threshold).astype(int)
```

```python
x=np.array([0.5,1,1.5,2,2.5,3]);
y=np.array([0,0.2,0.1,0.5,0.7,1]);
b0,b1=logistic_regression_fit(x,y)
print("Beta0:",b0,"Beta1:",b1);
y_pred = predict(x,b0,b1)
print("Predicted output:",y_pred)
acc = np.mean(y_pred == y)
print("Accuracy:",acc)
```

```python
plot.scatter(x,y)
plot.plot(x,y_pred)
```
