```python
#IMPLEMENTING THE LINEAR REGRESSION FROM SCRATCH.
import numpy as np;
```

```python
hours=np.array([1,2,3,4,5]);
marks=np.array([40,50,60,70,80]);
print("Hours:",hours,"\nMarks:",marks);
```

```python
def linearRegression_fit(X,Y):
    n=len(X);
    sumx = np.sum(X);
    sumy = np.sum(Y);
    sumxy = np.sum(X*Y);
    nominator = n*sumxy - sumx*sumy;
    sumXsq = np.sum(X**2);
    sqSumX = np.pow(sumx,2);
    denominator = n*sumXsq - sqSumX
    m =nominator/denominator
    return m;
print(linearRegression_fit(hours,marks));
```
