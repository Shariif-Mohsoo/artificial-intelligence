```python
#implementation of linear regression from scratch
import numpy as np;
import matplotlib.pyplot as plot
```

```python
def linear_regression_fit(x,y):
    n=len(x)
    sumX=np.sum(x);
    sumY=np.sum(y);
    sumXY=np.sum(x*y)
    numerator = n*sumXY - (sumX * sumY)
    denominator = n*np.sum(x**2) - sumX**2
    B1 = numerator/denominator
    meanY=np.mean(y)
    meanX=np.mean(x)
    B0=meanY - B1*meanX
    return B0,B1
```

```python
def predict(B0,B1,X):
    return B0 + B1*X
```

```python
def mse(y_true,y_pred):
    return np.mean((y_true - y_pred)**2) 
```

```python
x=np.array([1,2,3,4,5]);
y=np.array([20,40,60,80,99]);
b0,b1=linear_regression_fit(x,y)
print("B0:",b0,"B1:",b1)
y_pred = predict(b0,b1,x)
print("Y_pred:",y_pred)
err=mse(y,y_pred)
print("MSE:",err)
```

```python
plot.scatter(x,y);
plot.plot(x,y_pred);
```

```python
#Implementing linear regression using scikit-learn
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
iris=load_iris()
# print(iris)
x=iris.data[:,:1];#sepal length as feature
y=iris.data[:,2] #petal length as target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model = LinearRegression()
model.fit(x_train,y_train)
y_pred1 = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred1)
print("MSE:",mse);
```

```python
plot.scatter(x,y)
plot.plot(x_test,y_pred1)
```
