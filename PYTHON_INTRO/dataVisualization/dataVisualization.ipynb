```python
import numpy as np;
import matplotlib.pyplot as plt;
import seaborn as sns;
import pandas as pd;

df = pd.read_csv("../Datasets /Iris.csv");
print(df)
```

```python
df.head()
```

```python
df.tail()
```

```python
#first let's explore little bit about scatter plot
#show's the relationship between two features
scatter_plot= sns.scatterplot(data=df,x="SepalLengthCm",y="SepalWidthCm",hue="Species");
scatter_plot.set_title("Iris Sepal Dimensions by species");
```

```python
# BarChart used to show count or comparison.
species_count = df["Species"].value_counts();
print(species_count)
plt.bar(species_count.index,species_count.values);
plt.xlabel("Species")
plt.ylabel("Count")
```

```python
#histogram shows how data is spread out.
plt.hist(df["PetalLengthCm"],bins=20,color="green");
```

```python
#line plot shows how something changes overtime.
sns.lineplot(data=df,x="SepalLengthCm",y="PetalLengthCm",hue="Species")
```

```python
#heatmap shows how strongly two variables are related.
corr= df.drop("Species",axis=1).corr();
print(corr)
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm");
```
