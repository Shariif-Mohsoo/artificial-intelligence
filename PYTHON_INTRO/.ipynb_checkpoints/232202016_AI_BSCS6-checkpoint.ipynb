```python
import numpy as np
math = np.array([95, 78, 64, 88, 72, 55, 90, 82, 100, 69])
science = np.array([88, 74, 70, 91, 66, 59, 85, 79, 96, 71])
english = np.array([92, 81, 68, 85, 77, 60, 89, 73, 94, 70])
```

```python
avg_math = np.mean(math)
avg_science = np.mean(science)
avg_english = np.mean(english)
print(f"Step 1 - Averages:\nMath: {avg_math:.2f}, Science: {avg_science:.2f}, English: {avg_english:.2f}\n")
max_math, min_math = np.max(math), np.min(math)
max_science, min_science = np.max(science), np.min(science)
max_english, min_english = np.max(english), np.min(english)
print(f"Step 2 - Max and Min:\nMath → Max: {max_math}, Min: {min_math}\nScience → Max: {max_science}, Min: {min_science}\nEnglish → Max: {max_english}, Min: {min_english}\n")
```

```python
pass_math = np.sum(math >= 70)
fail_math = len(math) - pass_math
pass_science = np.sum(science >= 70)
fail_science = len(science) - pass_science
pass_english = np.sum(english >= 70)
fail_english = len(english) - pass_english


print(f"Step 3 - Pass/Fail Counts per Subject:")
print(f"Math → Passed: {pass_math}, Failed: {fail_math}")
print(f"Science → Passed: {pass_science}, Failed: {fail_science}")
print(f"English → Passed: {pass_english}, Failed: {fail_english}\n")

```

```python
passed_all = np.sum((math >= 70) & (science >= 70) & (english >= 70))
failed_overall = len(math) - passed_all
print(f"Step 4 - Overall Passing:")
print(f"Students Passed All Subjects: {passed_all}")
print(f"Students Failed Overall: {failed_overall}\n")

student_avg = (math + science + english) / 3

def grade_letter(avg):
    if avg >= 90: return 'A'
    elif avg >= 80: return 'B'
    elif avg >= 70: return 'C'
    else: return 'F'
performance = np.array([grade_letter(a) for a in student_avg])

print("Step 5 - Student Grades:")
for i in range(len(math)):
    print(f"Student_{i+1}: Avg = {student_avg[i]:.2f}, Grade = {performance[i]}")
```
